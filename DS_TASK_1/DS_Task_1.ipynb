```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
os.listdir()
```

```python
os.listdir("API_SP.POP.TOTL_DS2_en_csv_v2_40826")
```

```python
file_path = "API_SP.POP.TOTL_DS2_en_csv_v2_40826.csv"
```

```python
df = pd.read_csv(file_path, skiprows=4)
```

```python
df.head()
```

```python
world_data = df[df["Country Name"] == "World"]
```

```python
year_columns = [col for col in world_data.columns if col.isdigit()]
```

```python
year_columns = [year for year in year_columns if int(year) <= 2026]
```

```python
population = world_data[year_columns].T
population = population.rename(columns={population.columns[0]: "Population"})
population.index = population.index.astype(int)
population = population.dropna()
```

```python
plt.figure(figsize=(14,6))
plt.bar(population.index, population["Population"])
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("World Population Over Years (Bar Chart)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
print(population.tail())
```

```python
plt.figure(figsize=(8,5))
plt.hist(population["Population"], bins=20)
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.title("Distribution of World Population Values (Histogram)")
plt.show()
```
